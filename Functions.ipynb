Question 1 :- What is the difference between a function and a method in Python

Answer :- In Python, both functions and methods are used to perform tasks and organize code, but they are used differently. A function is a block of code that is written using the def keyword and works independently. It is not connected to any object or class. Functions are useful when we want to do a task that doesn't depend on any object.

 For example, if we want to greet someone, we can create a function like def greet(name): return "Hello " + name. When we call greet("Vanshika"), it gives the output "Hello Vanshika". This function works on its own and doesn't need any class or object to run.

On the other hand, a method is also a function but it is defined inside a class and works only with objects. It always takes at least one parameter, usually called self, which refers to the object calling the method. Methods are used when we want to perform actions related to an object.

 For example, if we create a class Person, and inside it we write

 def say_hello(self, name):
  return "Hi " + name,
  
   then we must create an object like p = Person() and call p.say_hello("Vanshika") to get the output. This is because the method is tied to the object p.

In short, the main difference is that a function stands alone, while a method is always connected to a class and called through an object. Both are important in Python and help make our programs clean and organized.

Question 2 :-  Explain the concept of function arguments and parameters in Python

Answer :- In Python, parameters and arguments are important parts of how functions work. When we create a function, we define it using the def keyword, and inside the parentheses, we put parameters. These parameters act like placeholders for the values the function will use. For example, in def greet(name):, the word name is a parameter. It tells the function that when someone calls it, they should give a value for name.

Now, when we actually call the function, we provide real values for those parameters. These real values are called arguments.

For example, if we call greet("Vanshika"), the string "Vanshika" is an argument. So basically, parameters are used when we define a function, and arguments are used when we call the function.

Python allows different types of arguments:

-> Positional arguments, where the order matters.

-> Keyword arguments, where we mention the parameter name like greet(name="Vanshika").

-> Default arguments, where we give a default value to a parameter like def greet(name="Guest"), so if we don't pass anything, it uses “Guest”.

-> Arbitrary arguments, where we don't know how many arguments will be passed, so we use *args for multiple values or **kwargs for keyword arguments.

Question 5 :-  What are iterators in Python and how do they differ from iterables

Answer :- In Python, **iterators** and **iterables** are both used to go through a collection of items like lists, strings, or tuples, but they are not the same. An **iterable** is any object that can be looped over using a `for` loop. Common examples of iterables include lists, strings, tuples, sets, and dictionaries. These objects contain multiple elements and can give you one item at a time.

 For example:

my_list = [1, 2, 3]

for item in my_list:

    print(item)


Here, `my_list` is an **iterable**.

An **iterator**, on the other hand, is an object that **remembers its position** while going through the elements one by one. You get an iterator from an iterable by using the `iter()` function. Once you have an iterator, you can use the `next()` function to get the next item.

For example:


my_list = [1, 2, 3]

it = iter(my_list)


print(next(it))


print(next(it))  


This goes item by item until all are used. After that, it raises a `StopIteration` error.

The key difference is that **iterables can be looped over**, but **iterators are used to fetch items one at a time**. Iterators are useful when you don’t want to load all data at once, especially with large data sets. In short, **every iterator is an iterable**, but **not every iterable is an iterator** unless we use `iter()` on it.


Question 6 :-  Explain the concept of generators in Python and how they are defined

Answer :- In Python, **generators** are a special type of function that **generate values one at a time** using the `yield` keyword instead of `return`. They are used to create **iterators** in an easy and memory-efficient way, especially when working with large data sets. Unlike normal functions that return a single value and end, a generator **pauses** at each `yield` and continues from where it left off the next time it's called.

To define a generator, we use the `def` keyword like a normal function, but instead of `return`, we use `yield`. This makes the function a generator.

 example:

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
```

When we call `count_up_to(3)`, it doesn't run the whole function at once. Instead, it returns a generator object. We can loop through it using a `for` loop or call `next()` to get each value:

```python
gen = count_up_to(3)
print(next(gen))  
print(next(gen))  
print(next(gen))  
```

After all values are yielded, calling `next()` again will raise a StopIteration error.




Question 7 :-  What are the advantages of using generators over regular functions

Answer :- Generators have several advantages over regular functions, especially when dealing with large amounts of data. The main benefit is that generators do not store all values in memory. Instead, they produce one value at a time using the `yield` keyword. This makes them memory-efficient and useful for tasks like reading large files or working with big datasets.

For example, a regular function that returns a list of numbers from 1 to 1000000 will store all the numbers in memory, which can slow down the program. But a generator will give one number at a time, saving memory and improving speed.

```python
def generate_numbers(n):
    for i in range(n):
        yield i
```

Generators are also faster for loops and can handle infinite sequences without crashing, unlike regular functions that return a list. In simple words, generators are best when we want to loop through a lot of data without using too much memory.

Question 9 :-  Explain the purpose and usage of the `map()` function in Python

Answer :- The `map()` function in Python is used to apply a function to every item in a list (or any iterable) and return a new map object with the results. It is helpful when you want to perform the same operation on each element without using a loop.

The syntax is:

```python
map(function, iterable)
```

The function can be a normal function or a lambda function. The result is a map object, so we usually convert it to a list using `list()` to see the output.

 Example:

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x * x, numbers)
print(list(squared))
```

In this example, the lambda function `x * x` is applied to each item in the list `numbers`. The `map()` function makes the code shorter and faster when working with lists and other iterables.


Question 10 :-  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python

Answer :- In Python, `map()`, `reduce()`, and `filter()` are used to process data in a list or any iterable, but they do different things. The `map()` function applies a function to each item in the iterable and returns a new iterable with the results. For example:

```python
numbers = [1, 2, 3]
result = list(map(lambda x: x * 2, numbers))  
output ->[2, 4, 6]
```

The `filter()` function is used to select items that meet a condition. It returns only the items where the function returns `True`. Example:

```python
numbers = [1, 2, 3, 4]
result = list(filter(lambda x: x % 2 == 0, numbers))  
output -> [2, 4]
```

The `reduce()` function, from the `functools` module, reduces the list to a single value by applying the function repeatedly. Example:

```python
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
output -> 10
```

So, `map()` transforms, `filter()` selects, and `reduce()` combines.

````python
# Question 4 :-  Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


is_prime(23)
````

````python
# Question 5 :-  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms

class Fibonacci:
    def __init__(self, limit):
        self.a, self.b = 0, 1
        self.limit = limit
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        self.count += 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

fib = Fibonacci(6)

for num in fib:
    print(num)
````

````python
# Question 6 :-  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for power in powers_of_two(4):
    print(power)
````

````python
# Question 7 :-  Implement a generator function that reads a file line by line and yields each line as a string.

def read_lines(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip()
````

````python
# Question 10 :-  Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))

remove_vowels("python")
````
